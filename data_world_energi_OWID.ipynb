```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/owid/energy-data/master/owid-energy-data.csv"

df = pd.read_csv(url)

print(df.shape)
df.head()
```

```python
## Filter data only using World
df = df[df['country'] == 'World']
## The data that are chosen are from 1960
df = df[df['year'] >= 1990]
df.shape
```

```python
list_columns = df.columns.tolist()
list_columns
```

```python
cols = [
    "year", "primary_energy_consumption", "fossil_fuel_consumption", "renewables_consumption",
    "coal_consumption", "oil_consumption", "gas_consumption", "low_carbon_consumption", "nuclear_consumption",
    "coal_share_energy", "fossil_share_energy", "gas_share_energy", "low_carbon_share_energy", "nuclear_share_energy",
    "oil_share_energy", "renewables_share_energy",
]

df = df[cols].reset_index(drop=True)
df
```

```python
df['renewables_yoy_growth_pct'] = df['renewables_consumption'].pct_change() * 100
df['primary_energy_yoy_growth_pct'] = df['primary_energy_consumption'].pct_change() * 100
```

```python
df.to_csv("world_consumption_energy.csv")
```
